```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Simulated stock price data
np.random.seed(42)
days = np.arange(100).reshape(-1, 1)
price = 100 + 0.5 * days + np.random.normal(0, 5, size=(100, 1))

# Fit linear regression model
model = LinearRegression()
model.fit(days, price)
predictions = model.predict(days)

# Plot actual vs predicted prices
plt.figure(figsize=(8, 5))
plt.scatter(days, price, color='blue', label='Actual Price')
plt.plot(days, predictions, color='red', label='Predicted Trend')
plt.xlabel('Days')
plt.ylabel('Stock Price ($)')
plt.title('Linear Regression for Stock Price Prediction')
plt.legend()
plt.show()
```
